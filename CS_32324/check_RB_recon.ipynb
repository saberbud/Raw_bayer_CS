```python
import numpy as np
import h5py
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
from PIL import Image
from skimage.measure import compare_ssim,compare_psnr
```

```python
f_i=h5py.File('./E3/image_orig.h5','r')
image_orig=f_i['data'][0,:,:,0]
print("image_orig shape: " + str(image_orig.shape))
```

```python
plt.figure(figsize=(14,10))
plt.imshow(image_orig, cmap='gray')
```

```python
f_i=h5py.File('./E3/image_32324_rec_CI.h5','r')
image_rec=f_i['data'][0,:,:,0]
print("image_rec shape: " + str(image_rec.shape))
```

```python
plt.figure(figsize=(14,10))
plt.imshow(image_rec, cmap='gray')
```

```python
#Convert to uint8 and compare PSNR, SSIM
output=(image_rec*255).astype('uint8')
label=(image_orig*255).astype('uint8')

psnr =compare_psnr(output, label)
ssim =compare_ssim(output, label, multichannel=False)

print("psnr= " + str(psnr) + " , ssim= " + str(ssim))
```
